```r
library(tidyverse)
library(stringr)
```

```r
df = read_csv('/root/aics/modeling/gregj/results/integrated_cell/test_aaegan/aaegan3Dv8_v1-exp_v2/analysis/inception_scare/inception.csv')
colnames(df) = colnames(df)[c(3,2,1,4,5,6)]
```

```r
df %>%
    ggplot(aes(x=KLdiv_data, color=structureProteinName, fill=structureProteinName)) +
    geom_density(alpha=0.05, bw=0.1) +
    geom_rug(alpha=0.05) +
    theme_minimal() +
#     xlim(c(0,1)) +
    facet_wrap(~phase, scales = 'fixed') +
    theme(aspect.ratio=1, panel.spacing.x=unit(1, "lines")) +
    labs(x="KL Divergence", y= "Probability Density", color="Label", fill="Label")
```
